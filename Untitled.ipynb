```python
from sentiment_analysis.model import Model
from sentiment_analysis.common.checkpointer import Checkpointer
from sentiment_analysis.experiment import load_settings
from flax import nnx
from jax import numpy as jnp
import jax
from penzai import pz
import matplotlib.pyplot as plt
import seaborn as sns
import cmocean
from pathlib import Path

pz.ts.register_autovisualize_magic()
```

```python
path = Path("results/small_mixed_single_2024-07-18_13-20-12")
settings = load_settings(path / "settings.json")

original = Model(settings.model, nnx.Rngs(0))

checkpoints = Checkpointer(path / "checkpoints")
model = checkpoints.restore(original, 49999)
```

```python
state = nnx.state(model, nnx.Param)
position_embedding = state.position_embedding.embedding.value
pz.ts.render_array(position_embedding)
```

```python

def eval_model(model, tokens, rngs):
    return model(tokens, True, rngs) 
```

```python
tokens = vocab.tokenize("My partner got me these flowers on Sunday July 14. They were very pretty and unique; however, it is now Thursday and they are completely dead. They already started looking bad on Tuesday July 16. Price was 50 bucks.")
tokens = list(tokens)

tokens_str = []
for token in tokens:
    tokens_str.append(vocab.id_to_token(token))

print(tokens)
print(tokens_str)

padded_tokens = jnp.array(tokens + [-1] * (115 - len(tokens)), jnp.int16)
print(len(tokens))
```

```python
out = eval_model(model, padded_tokens, nnx.Rngs(0))
print(out)
print(nnx.softmax(out))
print(jnp.argmax(out) + 1)
```

```python
size = 1 + len(tokens)
pz.ts.render_array(model.transformer_layers[3].attention.attention_weights.value[0][:, :size, :size], truncate=False)
```

```python
attention = model.transformer_layers[0].attention.attention_weights.value[0]
combined = jnp.hstack(attention)
sns.heatmap(jnp.asarray(combined, dtype=jnp.float32), vmin=0, vmax=0.25, cmap=cmocean.cm.dense, square=True, cbar=False, xticklabels=False, yticklabels=False)
#nnx.display(attention)
```

```python
def heatmap(data, count):
    sns.heatmap(jnp.asarray(data, dtype=jnp.float32), vmin=0.0, vmax=0.15, cmap=cmocean.cm.dense, square=True, cbar=False, xticklabels=False, yticklabels=False)
    plt.savefig(f"render/attention_{count}.png", format="png", bbox_inches='tight')
```
